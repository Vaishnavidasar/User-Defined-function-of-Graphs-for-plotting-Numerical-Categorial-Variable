## User Defined function of Graphs for plotting Numerical & Categorial Variable

**EXPLANATION:**<br>
We need a function which will give histogram and boxplot for numerical data and bar plot for categorical data<br>
We need graphs for machine learning to make predictions, or discover new patterns in the given data.<br>
1. Define a function which takes data,columns for which graphs we want and the directory we want to save these graphs or else it will save in the working directory.
2. If we give columns as None or not given it will take as cols which is for all columns in the data.
3. If we want to change the directory where we want to save the graphs use os.chdir to change the directory.
4. Check if the given column is having categorical data or numerical data,for numerical data use data._get_numeric_data().columns and create a list of those columns using tolist().
5. For the input column, if it is in the numerical columns list it will perform histogram and boxplot.
6. For getting Boxplot use data.boxplot where notch is true, grid is false to remove grid lines ,figsize for size of figure and vert false to get horizontal plot.
7. For getting Histogram use data.hist where grid is false to remove grid lines and figsize for size of figure and bins for number of bins.
8. Give Xlabel as column name,ylabel as frequency and title to boxplot and histogram using plt.xlabel,plt.ylabel,plt.title where fontsize is 12.
9. Save the figure using plt.savefig where bbox_inches for the fit,pad_inches for padding around the figure,Transparent is False to get solid background and dpi for number of dots per inch.
10. plt.show() to display the graph.
11. Now else where the data is categorical and we should plot bar plot.
12. For bar plot count the values of the column using data.value_counts().plot where kind for type of plot(bar for barplot) and figsize for size of the figure ,color and fontsize for the size of the font.
13. Give Xlabel as column name,ylabel as frequency and title to bar plot using plt.xlabel,plt.ylabel,plt.title where fontsize is 12.
14. Save the figure using plt.savefig where bbox_inches for the fit,pad_inches for padding around the figure,Transparent is False to get solid background and dpi for number of dots per inch.
15. plt.show() to display the graph.

```python
def graphs(data,columns,directory=os.getcwd()):
    cols=data.columns
    if columns==None:
        columns=cols
    path = os.chdir(directory)
    num_cols = data._get_numeric_data().columns
    num_cols = num_cols.tolist()
    for i in columns:
        if (i in num_cols):               
                data.boxplot(column=i,notch =True,grid=False,figsize=(6,4),vert=False)
                plt.xlabel(i,fontsize=12)
                plt.ylabel("Frequency",fontsize=12)
                str1 = "Boxplot of "+i
                plt.title(str1,fontsize=14)
                plt.savefig(str1,bbox_inches='tight',pad_inches=2,transparent=False,dpi=100)
                plt.show()
                plt.draw()
                data.hist(i,grid = False,figsize=(6,4),bins=10)
                plt.xlabel(i,fontsize=12)
                plt.ylabel("Frequency",fontsize=12)
                str3 = "Histogram of "+i
                plt.title(str3,fontsize=14)
                plt.savefig(str3,bbox_inches='tight',pad_inches=2,transparent=False,dpi=100)
                plt.show()
        else:
                data[i].value_counts().plot(kind='bar',figsize=(6,6),color="coral",fontsize=8)
                plt.xlabel(i,fontsize=10)
                plt.ylabel("Frequency",fontsize=12)
                str5 = "Bar Plot of "+i
                plt.title(str5,fontsize=14)
                plt.savefig(str5,bbox_inches='tight',pad_inches=2,transparent=False,dpi=100)
                plt.show()
```

```python
#Plot graphs for cars data, given coluns and in the working  directory.
graphs(cars,['MPG','Origin'])
```
